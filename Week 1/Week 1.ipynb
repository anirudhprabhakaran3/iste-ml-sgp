```python
import pandas as pd

path = 'Titanic_ISTE.csv'
df = pd.read_csv(path)
```

```python
df.head()
```

```python
target = df["Survived"]

column1 = df["Pclass"]
column2 = df["Age"]

correlation1 = target.corr(column1)
correlation2 = target.corr(column2)

print("Correlation 1 is: ", correlation1)
print("Correlation 2 is: ", correlation2)
```

```python
column = df['Fare']
correlation = target.corr(column)
print("Correlation is: ", correlation)
```

```python
CrosstabResult1 = pd.crosstab(index=df['Pclass'], columns=df['Survived'])
CrosstabResult2 = pd.crosstab(index=df['Age'], columns=df['Survived'])

print(CrosstabResult1)
print(CrosstabResult2)
```

```python
from scipy.stats import chi2_contingency

ChiSqResult1 = chi2_contingency(CrosstabResult1)
ChiSqResult2 = chi2_contingency(CrosstabResult2)

print("P value for ChiSq test 1 is: ", ChiSqResult1[1])
print("P value for ChiSq test 2 is: ", ChiSqResult2[1])
```
